```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 2. 载入鸢尾花数据
iris = load_iris()
X = iris.data          # (150,4)  4 个特征
y = iris.target        # (150,)   3 个类别 0/1/2
target_names  = iris.target_names         # ['setosa' 'versicolor' 'virginica']

target_names
```

```python
# 3. 标准化
X_std = StandardScaler().fit_transform(X)

# 4. PCA 降到 2 维
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X_std)          # (150,2)

# 5. 把降维后的结果整理成 DataFrame
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
df_pca['Species'] = [target_names[i] for i in y]

df_pca
```

```python
# 6. 各主成分解释方差比例
print("各主成分解释方差比例:", pca.explained_variance_ratio_)
print("累计解释方差比例:", pca.explained_variance_ratio_.cumsum())
```
